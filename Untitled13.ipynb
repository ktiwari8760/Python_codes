```python
import pandas as pd
header = ["id" , "name" , "age"]
data = list(map(lambda x : x.split(',') , ["1 , kartik , 12" , "2 , rahul , 32" , "3 , chandu , 56" , "4 , prateek , 12" , "5 , keshav , 21" ]))

def func(data):
  list1 = []
  for ele in data:
    dict2 = {}
    dict2["id"] = int(ele[0])
    dict2["name"] = ele[1]
    dict2["age"] = float(ele[2])
    list1.append(dict2)
  return list1
dict3 = func(data)
#print(dict3)
df = pd.DataFrame(dict3)
df.head()
df.describe()
df1 = df['age'].min()
df2 = df[df["age"] == 12.0]
print(df1)
df.info()
```
